```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = 'GBOOST'
param = 'CVI'
path = '/content/drive/Shareddrives/Unlimited Drive | @LicenseMarket/subset1/S2_'+ param +'/'+ model
cf_matrix = np.load(path + '/conMatrix.npy')
```

```python
import seaborn as sns
plt.figure()

ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues',fmt='')

ax.set_title(model + ' crop mapping with ' + param + ' confusion matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Wheat','Vegetation','Barely','Colza','bare'])
ax.yaxis.set_ticklabels(['Wheat','Vegetation','Barely','Colza','bare'],rotation='horizontal')
plt.show()
## Display the visualization of the Confusion Matrix.
plt.savefig(path+'/conf.png', dpi = 300, bbox_inches='tight', pad_inches=0.01)
```
